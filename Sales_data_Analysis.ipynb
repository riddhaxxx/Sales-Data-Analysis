```python
#Import Libraries
from IPython.core import prefilter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv("superstore.csv.csv")
df.head()
df.info()
df.describe()
```

```python
#Data Cleaning
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Order.Date"] = pd.to_datetime(df["Order.Date"])
```

```python
#Data Analysis
df['Sales'].sum()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
#Data Visualization
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel('Categroy')
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby('Year')['Sales'].sum().plot(kind='line')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
#Insights

# 1. The Technology category generated the highest sales, indicating strong demand.
# 2. Profit varies across regions, showing differences in business performance.
# 3. Some categories have high sales but lower profit, suggesting cost or discount impact.
# 4. Sales trends fluctuate over time, indicating possible seasonal demand patterns.
# 5. Understanding both sales and profit is important for better decision-making
```
